```python
import time
import random


TESTS_AMOUNT = 1
ARRAY_SIZE = 100_000_000


def tester(func, *args):
    t = time.time()
    func(*args)
    return time.time() - t


def binary_search(sorted_array, k):
    if len(sorted_array) == 0:
        return False
    m = len(sorted_array) // 2
    if sorted_array[m] > k:
        return binary_search(sorted_array[:m], k)
    elif sorted_array[m] < k:
        return binary_search(sorted_array[m+1:], k)
    else:
        return True

# Iterative Binary Search Function 
# It returns location of x in given array arr if present, 
# else returns -1 
def binarySearch(arr, l, r, x): 
    
    while l <= r: 
  
        mid = l + (r - l)//2; 
          
        # Check if x is present at mid 
        if arr[mid] == x: 
            return mid 
  
        # If x is greater, ignore left half 
        elif arr[mid] < x: 
            l = mid + 1
  
        # If x is smaller, ignore right half 
        else: 
            r = mid - 1
      
    # If we reach here, then the element was not present 
    return -1

    
def ternary_search(sorted_array, k):
    if len(sorted_array) == 0:
        return False
    l = len(sorted_array) // 3
    r = l * 2
    if sorted_array[l] > k:
        return ternary_search(sorted_array[:l], k)
    elif sorted_array[l] == k:
        return True
    elif sorted_array[r] > k:
        return ternary_search(sorted_array[l+1:r], k)
    elif sorted_array[r] == k:
        return True
    else:
        return ternary_search(sorted_array[r+1:], k)

    
# Python 3 program to illustrate iterative 
# approach to ternary search 
  
# Function to perform Ternary Search 
def ternarySearch(ar, l, r, key): 
    while r >= l: 
          
        # Find mid1 and mid2 
        mid1 = l + (r-l) // 3
        mid2 = r - (r-l) // 3
  
        # Check if key is at any mid 
        if key == ar[mid1]: 
            return mid1 
        if key == mid2: 
            return mid2 
  
        # Since key is not present at mid,  
        # Check in which region it is present 
        # Then repeat the search operation in that region 
        if key < ar[mid1]: 
            # key lies between l and mid1 
            r = mid1 - 1
        elif key > ar[mid2]: 
            # key lies between mid2 and r 
            l = mid2 + 1
        else: 
            # key lies between mid1 and mid2 
            l = mid1 + 1
            r = mid2 - 1
  
    # key not found 
    return -1
    
t1 = 0
t2 = 0
ra = list(range(ARRAY_SIZE))
for i in range(TESTS_AMOUNT):
    print(f'[*] Test {i}...')
#     ra = sorted([random.randint(0, 2 * ARRAY_SIZE) for _ in range(ARRAY_SIZE)])
#     tmp = random.randint(0, 2 * ARRAY_SIZE)
    
    t1 += tester(binarySearch, ra, 0, ARRAY_SIZE-1, ra[-1])
    t2 += tester(ternarySearch, ra, 0, ARRAY_SIZE-1, ra[ARRAY_SIZE // 2])

t1 /= TESTS_AMOUNT
t2 /= TESTS_AMOUNT
print(t1, t2)
```
